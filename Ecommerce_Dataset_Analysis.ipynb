## **Ecommerce Dataset Analysis**

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Ecom = pd.read_csv("/content/ecommerce_dataset_10000.csv")
```

```python
Ecom.head()
```

```python
Ecom.describe()
```

```python
# Shape of Data

Ecom.shape
```

```python
# Data information
Ecom.info()
```

```python
# Finding null values
Ecom.isnull().sum()
```

```python
# Created a new column of full name

Ecom["Full_Name"] = Ecom["first_name"]+" "+Ecom["last_name"]
```

```python
Ecom.head()
```

```python
# Created a new column of Total ammount fror total sales

Ecom["Total_Ammount"] = Ecom["quantity"]*Ecom["unit_price"]
```

```python
Ecom.head()
```

```python
print(Ecom.columns.tolist())
```

```python
Ecom.rename(columns={'customer_id\t':"customer_id"},inplace = True)
```

**Customer** **Analysis**

```python
# customers by Gender

Ecom.groupby("gender")["customer_id"].nunique().sort_values(ascending= False)
```

```python
# customers by age group
Ecom.groupby("age_group")["customer_id"].nunique().sort_values(ascending = False)
```

```python
# customers by country
Ecom.groupby("country")["customer_id"].nunique().sort_values(ascending = False)
```

Orders and Sales Analysis

```python
# Most orders received from which country

Ecom["country"].value_counts()
```

```python
# Order status
Ecom["order_status"].value_counts()
```

This shows that many orders are cancelled or pending or returned

```python
# Most orders received by which age group

Ecom["age_group"].value_counts()
```

```python
Ecom.head()
```

```python
# Age group wise sales
Age_group_sales = Ecom.groupby("age_group")["Total_Ammount"].sum().sort_values(ascending = False)
Age_group_sales
```

```python
plt.bar(Age_group_sales.index,Age_group_sales.values)
plt.xlabel("Age groups")
plt.ylabel("Total_Ammount")
plt.title("Age group wise sales")
plt.ticklabel_format(style='plain', axis='y')
```

```python
Ecom.head()
```

```python
# Countries wise total sales

countries_wise_sales = Ecom.groupby("country")["Total_Ammount"].sum().sort_values(ascending = False)
countries_wise_sales
```

```python
plt.barh(countries_wise_sales.index,countries_wise_sales.values)
plt.xlabel("Countries")
plt.ylabel("Total Sales")
plt.title("Countries wise total sales")
```

```python
# Top 5 countries sales wise

Top5 = Ecom.groupby("country")["Total_Ammount"].sum().sort_values(ascending = False).head(5)
Top5
```

```python
plt.bar(Top5.index,Top5.values)
plt.xlabel("Countries")
plt.ylabel("Total Sales")
plt.title("Top 5 countries sales wise")
```

```python
Ecom.head()
```

```python
#  Gender - based Sales Analysis

Gender_wise = Ecom.groupby("gender")["Total_Ammount"].sum()
Gender_wise
```

```python
plt.pie(Gender_wise.values,labels = Gender_wise.index,autopct = "%1.2f%%")
plt.title("Sales Distribution by Gender")
```

```python
# Category-wise Sales breakdown

category_wise = Ecom.groupby("category")["Total_Ammount"].sum().sort_values(ascending = False)
category_wise
```

```python
plt.bar(category_wise.index,category_wise.values)
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel("Category")
plt.ylabel("Total sales")
plt.title("Category wise sales breakdown")
```

By doing category wise sales analysis we can say that ,Its evident that Electronics dominates the sales figures by a significant margin, contributing the highest revenue .

The remaining categories   Home - Kitchen , Apparel, Toys, Books, and Sports  show relatively lower but roughly similar sales volumes.

```python
# The total number of unique products sold or listed in the dataset.

Ecom["product_id"].unique().size
```

```python
# Product wise sales brakedown

Ecom.groupby("product_name")["Total_Ammount"].sum().sort_values(ascending = False)
```

```python
# Top 5 Product Sales wise

Top5product = Ecom.groupby("product_name")["Total_Ammount"].sum().sort_values(ascending = False).head(5)
Top5product
```

```python
plt.bar(Top5product.index , Top5product.values)
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.title("Top 5 products sales wise")
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()

```
